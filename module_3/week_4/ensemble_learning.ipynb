# **Ensemble Learning**

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib . pyplot as plt

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

## Load dataset

```python
dataset_path = "./Housing.csv"
df = pd.read_csv(dataset_path)
```

```python
df
```

## Process categorical

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.to_list()
```

```python
print(categorical_cols)
```

```python
ordinal_encoder = OrdinalEncoder()
encoded_categorical_cols = ordinal_encoder.fit_transform(df[categorical_cols])
encoded_categorical_df = pd.DataFrame(encoded_categorical_cols, columns=categorical_cols)
numerical_df = df.drop(categorical_cols, axis=1)
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis=1)
```

```python
encoded_df
```

## Normalize Dataset

```python
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)
```

```python
dataset_arr
```

## Split X, y data

```python
X, y = dataset_arr[:, 1:], dataset_arr[:, 0]
```

## Split train, val data

```python
test_size = 0.3
random_state = 1
is_shuffle = True
X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=test_size,
    random_state=random_state,
    shuffle=is_shuffle
)
```

## Train model

```python
# Random Forest model
rf_regressor = RandomForestRegressor(random_state=random_state)
rf_regressor.fit(X_train, y_train)
```

```python
# AdaBoost model
ab_regressor = AdaBoostRegressor(random_state=random_state)
ab_regressor.fit(X_train, y_train)
```

```python
# Gradient Boosting model
gb_regressor = GradientBoostingRegressor(random_state=random_state)
gb_regressor.fit(X_train, y_train)
```

## Evaluation

### Random Forest

```python
y_pred = rf_regressor.predict(X_val)
```

```python
# Use Mean Absolute Error (MAE) and Mean Squared Error (MSE)
mae = mean_absolute_error(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

print("Evaluation results on validation set: ")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

### AdaBoost

```python
y_pred = ab_regressor.predict(X_val)
```

```python
# Use Mean Absolute Error (MAE) and Mean Squared Error (MSE)
mae = mean_absolute_error(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

print("Evaluation results on validation set: ")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

### Gradient Boosting

```python
y_pred = ab_regressor.predict(X_val)
```

```python
# Use Mean Absolute Error (MAE) and Mean Squared Error (MSE)
mae = mean_absolute_error(y_val, y_pred)
mse = mean_squared_error(y_val, y_pred)

print("Evaluation results on validation set: ")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```
